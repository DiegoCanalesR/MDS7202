### Análisis Univariado  

Observación de las distribuciones de las variables involucradas, tanto numéricas como categóricas.

````python
numeric = ['math score', 'writing score', 'reading score']
categorical = [col for col in list(df_grades.columns) if col not in numeric]
````

````python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

#Análisis univariado de variables numéricas
fig = make_subplots(rows=1, cols=3, subplot_titles=numeric)

for idx, col in enumerate(numeric):
    row_idx = 1
    col_idx = idx+1

    data = df_grades.loc[:, col]
    hist = go.Histogram(x=data, name=col, histnorm="probability")
    fig.add_trace(hist, row=row_idx, col=col_idx)

fig.update_layout(
    height=400,
    title_text="Análisis Univariado de las Variables Numéricas",
    showlegend=False,
)
fig.show()


#Análisis univariado de variables categóricas
n_cols = 2
fig = make_subplots(rows=3, cols=n_cols, subplot_titles=categorical)

for idx, col in enumerate(categorical):
    row_idx = idx // n_cols + 1
    col_idx = idx % n_cols + 1

    data = df_grades.loc[:, col]
    hist = go.Histogram(x=data, name=col)
    fig.add_trace(hist, row=row_idx, col=col_idx)

fig.update_layout(
    height=1200,
    title_text="Análisis Univariado de las Variables Categoricas",
    showlegend=False,
)
fig.show()
````

**Hint**

En el caso que decida limpiar el dataset, puede que le sean de utilidad las funciones ```.drop_duplicates()``` y ```.dropna()```.


````python
def limpieza_datos(df):
    #Eliminar filas repetidas
    df = df.drop_duplicates()
    
    #Eliminar nombres repetidos
    df = df.drop_duplicates(subset=['names'])
    
    #Enmascarar valores fuera de rango con NaNs
    df['math score'] = df['math score'].mask(df['math score'] > 100)
    df['reading score'] = df['reading score'].mask(df['reading score'] > 100)
    df['writing score'] = df['writing score'].mask(df['writing score'] > 100)
    
    #ELiminar filas en que 3 scores son NaN
    df = df.dropna(axis=0, subset=numeric, how='all') 
    
    return df
````

> Utilice la función creada en el punto 1.1 para verificar sus resultados.

````python
df_grades_clean = limpieza_datos(df_grades)
analisis_preliminar_datos(df_grades_clean)
````

Se verifica que los valores máximos no exceden los rangos permitidos. También se observa que la canntidad de valores nulos se redujo y que ya no hay filas repetidas. También podemos verificar lo primero revisando nuevamente la distribución de las variables numéricas.

````python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

#Análisis univariado de variables numéricas
fig = make_subplots(rows=1, cols=3, subplot_titles=numeric)

for idx, col in enumerate(numeric):
    row_idx = 1
    col_idx = idx+1

    data = df_grades_clean.loc[:, col]
    hist = go.Histogram(x=data, name=col, histnorm="probability")
    fig.add_trace(hist, row=row_idx, col=col_idx)

fig.update_layout(
    height=400,
    title_text="Análisis Univariado de las Variables Numéricas",
    showlegend=False,
)
fig.show()
````

**Justificación:**

```
Las asunciones realizadas son las siguientes: primero, no se consideró válida la existencia de filas repetidas, ya que esto corresponde a información redundante que no es un aporte a los datos. En segundo lugar, se asume que cada alumno rinde una única vez cada prueba, por lo que no pueden haber filas con el mismo atributo 'names', porque esto entregaría múltiples puntajes en las prueba para cada nombre. También se asumió como válido el hecho de que un alumno pueda haber rendido 1, 2 o 3 pruebas, por lo que se admite la existencia de NaNs en alguna de las columnas numéricas. No obstante, se consideró inválido el caso en que los 3 valores son NaN, ya que este caso no aporta información al dataset, ya sea porque el alumno no rindió las pruebas o por errores en los registros. Al tratarse de rendimientos individuales de los alumnos, se decidió no imputar los valores con el promedio del resto (o alguna otra técnica) y se optó por eliminar dichas filas.
```

> Explique (pensando en una operación de conjuntos) que hace la operación de combinación por defecto (`inner`).

> Explique que hubiese sucedido si aplicaramos los siguientes merge en las variables anteriores:

- `left`
- `right`
- `outer`

**Justificación:**

```
La operación por defecto "inner" combina los elementos de ambos dataset conservando la intersección (pensándolo como conjuntos). Es decir, conserva los elementos que se encuentran en ambas tablas, descartando los que no encuentran su par. 

Si se aplica "left", además de la intersección se conservan las filas del primer dataset que no se combinaron con ninguna fila del segundo dataset.

Al aplicar "right" ocurre lo análogo a lo anterior, pero conservando las filas del segundo dataset.

Con "outer" ocurre que además de la intesección, se almacenan todas las filas de ambos dataset que no fueron combinadas.
```
