### An치lisis Univariado

````python
numeric = ['math score', 'writing score', 'reading score']
categorical = [col for col in list(df_grades.columns) if col not in numeric]
````

**Hint**

En el caso que decida limpiar el dataset, puede que le sean de utilidad las funciones ```.drop_duplicates()``` y ```.dropna()```.


````python
def limpieza_datos(df):
    #Eliminar filas repetidas
    df = df.drop_duplicates()
    
    #Eliminar nombres repetidos
    df = df.drop_duplicates(subset=['names'])
    
    #Enmascarar valores fuera de rango con NaNs
    df['math score'] = df['math score'].mask(df['math score'] > 100)
    df['reading score'] = df['reading score'].mask(df['reading score'] > 100)
    df['writing score'] = df['writing score'].mask(df['writing score'] > 100)
    
    #ELiminar filas en que 3 scores son NaN
    df = df.dropna(axis=0, subset=numeric, how='all') 
    
    return df
````

> Utilice la funci칩n creada en el punto 1.1 para verificar sus resultados.

````python
df_grades_clean = limpieza_datos(df_grades)
analisis_preliminar_datos(df_grades_clean)
````

Se verifica que los valores m치ximos no exceden los rangos permitidos. Tambi칠n se observa que la canntidad de valores nulos se redujo y que ya no hay filas repetidas. Tambi칠n podemos verificar lo primero revisando nuevamente la distribuci칩n de las variables num칠ricas.

````python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

#An치lisis univariado de variables num칠ricas
fig = make_subplots(rows=1, cols=3, subplot_titles=numeric)

for idx, col in enumerate(numeric):
    row_idx = 1
    col_idx = idx+1

    data = df_grades_clean.loc[:, col]
    hist = go.Histogram(x=data, name=col, histnorm="probability")
    fig.add_trace(hist, row=row_idx, col=col_idx)

fig.update_layout(
    height=400,
    title_text="An치lisis Univariado de las Variables Num칠ricas",
    showlegend=False,
)
fig.show()
````

En segundo lugar, se asume que cada alumno rinde una 칰nica vez cada prueba, por lo que no pueden haber filas con el mismo atributo 'names', porque esto entregar칤a m칰ltiples puntajes en las prueba para cada nombre. Al aplicar "right" ocurre lo an치logo a lo anterior, pero conservando las filas del segundo dataset.

```
Con "outer" ocurre que adem치s de la intesecci칩n, se almacenan todas las filas de ambos dataset que no fueron combinadas.
```
