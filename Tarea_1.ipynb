Para el caso de la codificaci√≥n c√≠clica deber√° crear una clase ```class CyclicalEncoding(BaseEstimator, TransformerMixin)``` para la codificaci√≥n c√≠clica, por lo que se recomienda ver la [documentaci√≥n](https://sklearn-template.readthedocs.io/en/latest/user_guide.html).

````python
class CyclicalEncoding(BaseEstimator, TransformerMixin):     
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        months_names = X.values
        if len(months_names.shape) > 1:
            datetime_objects = [datetime.datetime.strptime(x[0], "%B") for x in months_names]
        else:
            datetime_objects = [datetime.datetime.strptime(x, "%B") for x in months_names]
        months = np.array([date.month for date in datetime_objects])
        
        sin_ft = np.sin(2*np.pi*months/12)
        cos_ft = np.cos(2*np.pi*months/12)
        features = np.stack((sin_ft, cos_ft), axis=1)

        return features
````

````python
cyc = CyclicalEncoding()
cyc_ft = cyc.fit_transform(data.loc[:, 'Month'])
cyc_ft.shape
````

````python
class Identity(BaseEstimator, TransformerMixin):     
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        return X
````

### 2.3.2 Pipeline! [1 Punto]

Realizado el encoding a√±adan este a un pipeline, anexando como √∫ltimo paso un modelo de regresi√≥n. Para el modelo de regresi√≥n puede escoger uno de su preferencia o escoger un ```DecisionTreeRegressor()``` o ```LogisticRegression()```. 

Hecho esto, ```Kiguix``` en su rol de querer saber fehacientemente el desempe√±o de este modelo le sugiere utilizar R2 score, el cual es un estad√≠stico que nos permite visualizar que tan bien predice un modelo de regresi√≥n. ¬øQue tal le va a la regresi√≥n?, ¬øSera coherente utilizar solo estos atributos para predicciones?, Justifique su respuesta.

````python
X_train, X_test, y_train, y_test = train_test_split(
        data.drop(columns='Temp'),
        data.Temp,
        shuffle=True,
        test_size=0.3,
        random_state=42)

preprocessing_transformer = ColumnTransformer(
                            transformers=[
                                ('OneHotEncoder', OneHotEncoder(), ['station']),
                                ('MinMaxScaler', MinMaxScaler(), ['latitud', 'longitud','altura']),
                                ('CyclicalEncoder', CyclicalEncoding(), ['Month']),
                                ('Identity', Identity(), ['Year'])])

pipeline = Pipeline([('Preprocessing', preprocessing_transformer),
                    ('DecisionTreeRegressor', DecisionTreeRegressor())])

pipeline.fit(X_train, y_train)
predicted_regression = pipeline.predict(X_test)

# Metrica util para medir regresiones
R2score = r2_score(y_test, predicted_regression)
R2score
````

**Justifique:**

```
Teniendo en cuenta que el valor de R2 va entre 0 y 1, siendo 1 el mejor valor posible, se obtiene que la regresi√≥n muestra un desempe√±o considerablemente alto, tomando el valor aproximado de 0.92. Por lo anterior, se afirma que es razonable utilizar los atributos considerados para hacer la regresi√≥n, ya que los resultados obtenidos son suficientemente buenos seg√∫n el score calculado. No obstante, se hizo una prueba sin considerar los atributos "Year" y se obtiene un desempe√±o a√∫n mejor, logrando un R2 mayor que 0.95. 
```
